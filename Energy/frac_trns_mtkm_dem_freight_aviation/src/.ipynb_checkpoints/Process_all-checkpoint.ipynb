```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
```

```python
#df_freight=pd.read_csv("../row_data/DP_LIVE_18012023235438219.csv")
df_freight=pd.read_csv("../row_data/imputated_freight_data.csv")
df_aviation=pd.read_csv("../row_data/Aviation_freigth_inputed.csv",index_col=0)

df_freight=df_freight.fillna(0)
df_freight
```

```python
df_aviation
```

```python
import os

# set the directory path and the prefix string
directory = grandparentdir
prefix = 'frac_trns_mtkm_dem_freight'

frac_mtkm=[]
# iterate over all items in the directory
for item in os.listdir(directory):
    # check if the item is a directory and starts with the prefix
    if os.path.isdir(os.path.join(directory, item)) and item.startswith(prefix):
        frac_mtkm.append(item)
        # print the directory name
        print(item)
frac_mtkm
```

```python
nation_in_df=sorted(set(df_freight['LOCATION'].unique()).intersection(set(df_aviation['iso_code3'].unique())))
years_hist=np.arange(2011,2021)
df_all=df_aviation.copy()
for count in nation_in_df:
    for year in years_hist:
        try:
            RAIL_data=df_freight.loc[(df_freight['LOCATION']==count) &
                                 (df_freight['TIME']==year) &
                   (df_freight['SUBJECT']=='RAIL')
                  ]['Value'].values[0]
        except:

            RAIL_data=0

        try:
            ROAD_data=df_freight.loc[(df_freight['LOCATION']==count) &
                                 (df_freight['TIME']==year) &
                   (df_freight['SUBJECT']=='ROAD')
                  ]['Value'].values[0]
        except:
            ROAD_data=0

        try:
            INLAND_data=df_freight.loc[(df_freight['LOCATION']==count) &
                                 (df_freight['TIME']==year) &
                   (df_freight['SUBJECT']=='INLAND')
                  ]['Value'].values[0]
        except:
            INLAND_data=0

        try:
            COAST_data=df_freight.loc[(df_freight['LOCATION']==count) &
                                 (df_freight['TIME']==year) &
                   (df_freight['SUBJECT']=='COAST')
                  ]['Value'].values[0]
        except:
            COAST_data=0

        df_all.loc[(df_all['iso_code3']==count)&
               (df_all['Year']==year),'RAIL']=RAIL_data
        df_all.loc[(df_all['iso_code3']==count)&
               (df_all['Year']==year),'ROAD']=ROAD_data
        df_all.loc[(df_all['iso_code3']==count)&
               (df_all['Year']==year),'INLAND']=INLAND_data
        df_all.loc[(df_all['iso_code3']==count)&
               (df_all['Year']==year),'COAST']=COAST_data
    
df_all.dropna(axis=0, how='any',inplace=True)
```

```python
df_all['Total']=df_all.apply(lambda row: row['Aviation (mtkm)']+
                            row['RAIL']+row['ROAD']+row['INLAND']+row['COAST'],axis=1)
df_all
```

```python
# set the directory path and the prefix string
directory = grandparentdir
prefix = 'frac_trns_mtkm_dem_freight'

frac_mtkm=[]
# iterate over all items in the directory
for item in os.listdir(directory):
    # check if the item is a directory and starts with the prefix
    if os.path.isdir(os.path.join(directory, item)) and item.startswith(prefix):
        frac_mtkm.append(item)
        # print the directory name
        print(item)
frac_mtkm
```

```python
df_all['frac_trns_mtkm_dem_freight_water_borne']=df_all.apply(lambda row: 
    row['COAST']/row['Total'] if row['Total'] !=0 else 0 ,axis=1)
df_all['frac_trns_mtkm_dem_freight_road_heavy_freight']=df_all.apply(lambda row: 
    (row['ROAD']+row['INLAND'])/row['Total'] if row['Total'] !=0 else 0 ,axis=1)
df_all['frac_trns_mtkm_dem_freight_rail_freight']=df_all.apply(lambda row: 
    (row['RAIL'])/row['Total'] if row['Total'] !=0 else 0 ,axis=1)
df_all['frac_trns_mtkm_dem_freight_aviation']=df_all.apply(lambda row: 
    (row['Aviation (mtkm)'])/row['Total'] if row['Total'] !=0 else 0 ,axis=1)


df_all
```

```python
c_y=[]
for c in df_all['Nation'].unique():
    iso3=df_countries.loc[df_countries['Country Name']==c,'ISO3'].values[0]
    for y in np.arange(max(years_hist)+1,2051):
        c_y.append([c,iso3,y])

df_all_proj=pd.DataFrame(c_y,columns=['Nation','iso_code3','Year'])
        
df_all_proj
```

```python
for ssp_var in frac_mtkm:
    df_all_proj[ssp_var]=0
        
df_all_proj
```

```python
for ssp_var in frac_mtkm:
    for c in df_all['Nation'].unique():
        for y in np.arange(max(years_hist)+1,2051):
            df_all_proj.loc[(df_all_proj['Nation']==c)&
               (df_all_proj['Year']==y),ssp_var]=df_all.loc[(df_all['Nation']==c)&
               (df_all['Year']==max(years_hist))][ssp_var].values[0]
        
df_all_proj
```

```python
for ssp_var in frac_mtkm:
    df_all.to_csv(grandparentdir+'/'+ssp_var+"/input_to_sisepuede/historical/"+ssp_var+".csv")
    df_all_proj.to_csv(grandparentdir+'/'+ssp_var+"/input_to_sisepuede/projected/"+ssp_var+".csv")
```
