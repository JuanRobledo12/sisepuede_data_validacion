```python
# Add Parent Folder to System Path

import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir) 

####

import pandas as pd
import numpy as np

#Variable name
var_name='occrate_trns_passenger_per_vehicle_rail_passenger'
#DATAFRAMES
df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_data=pd.read_excel ('../row_data/occ_rate_trains_passenger.xlsx')
df_data
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=np.unique(df_data['Year'].values)
coun_year=[]
for count in zip(countries,ISO3):
    for year in years:
        coun_year.append([count[0],count[1],year])
df_historical_out = pd.DataFrame(coun_year, columns=['Country','iso_code3', 'Year'])
df_historical_out
```

```python
df_historical_out[var_name]=0
```

```python
#country_in_data=np.array([])
for count,year in zip(df_historical_out['Country'].values,df_historical_out['Year'].values):
    
    if count in df_data['Country'].values:
        df2=df_data[(df_data['Country'] == count) & (df_data['Year']== year)]
        data_value=df2['Oc_rate'].values
        df_historical_out.loc[(df_historical_out['Country'] == count) & 
                              (df_historical_out['Year']== year), var_name] = data_value
 #       country_in_data=np.append(country_in_data,1)
    else:
        df_historical_out.loc[(df_historical_out['Country'] == count) & 
                              (df_historical_out['Year']== year), var_name] = np.average(df_data['Oc_rate'].values)
  #      country_in_data=np.append(country_in_data,0)
#df_historical_out['country_in_data'] = country_in_data  
df_historical_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
#### PROJECTED ######

#Prediction for the next years will be the average value
years2=np.arange(min(years),2051,1)
coun_year=[]
for count in zip(countries,ISO3):
    for year in years2:
        coun_year.append([count[0],count[1],year])
df_projected_out = pd.DataFrame(coun_year, columns=['Country','iso_code3', 'Year'])
df_projected_out
```

```python
df_projected_out[var_name]=0
```

```python
for count,year in zip(df_projected_out['Country'].values,df_projected_out['Year'].values):
    if year in df_historical_out['Year'].values:
        df2=df_historical_out[(df_historical_out['Country'] == count) & 
                    (df_historical_out['Year']== year)]
        data_value=df2[var_name].values
        df_projected_out.loc[(df_projected_out['Country'] == count) & 
                              (df_projected_out['Year']== year), var_name] = data_value
    else:
        df_projected_out.loc[(df_projected_out['Country'] == count) & 
                              (df_projected_out['Year']== year), var_name] = np.average(df_data['Oc_rate'].values)
    
df_projected_out
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
