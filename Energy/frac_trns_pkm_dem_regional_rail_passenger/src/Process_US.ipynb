```python
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)

var_name=os.path.basename(parentdir)

var_name
```

```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
```

```python
#df_pass=pd.read_csv("../row_data/passenger-transport-volume-6.csv")
#df_freight=df_freight.fillna(0)
df_pass=pd.read_excel('../row_data/EPS_Transport_Data_-_Travel_Demand,_Fuel_Use,_and_Emissions_-_Travel_Demand_(Passenger_Modes).xlsx',
                      sheet_name='US Travel Demand')
df_pass.columns = df_pass.iloc[0]
df_pass
```

```python
df_pass['Transportation Mode Used on Trip'].values
```

```python
#Cross Walk with SISEPUEDE
Transport_cross_walk={
    'frac_trns_pkm_dem_private_and_public_human_powered':['Walk','Bicycle'],
    
    'frac_trns_pkm_dem_private_and_public_powered_bikes':['Motorcycle / Moped'],
    
    'frac_trns_pkm_dem_private_and_public_public':['School bus','Public or commuter bus',
                                                   'Subway / elevated / light rail / street car'],
    
    'frac_trns_pkm_dem_private_and_public_road_light':['Car','SUV', 'Van', 'Pickup truck', 
                                                       'Golf cart / Segway',
                                                       'RV (motor home, ATV, snowmobile)',
                                                       'Taxi / limo (including Uber / Lyft)',
                                                       'Rental car (Including Zipcar / Car2Go)'],
    
    'frac_trns_pkm_dem_regional_aviation':['Airplane'],
    'frac_trns_pkm_dem_regional_rail_passenger':['Amtrak / Commuter rail'],
    'frac_trns_pkm_dem_regional_road_heavy_regional':['City-to-city bus (Greyhound, Megabus)',
                                                     'Paratransit / Dial-a-ride',
                                                      'Private / Charter / Tour / Shuttle bus']

}

#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=range(2005,2018)

coun_year=[]
for count in zip(countries,ISO3):
    for year in years:
        coun_year.append([count[0],count[1],year])
df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
df_historical_out[var_name]=0
df_historical_out
```

```python
for count,year in zip(df_historical_out['iso_code3'].values,df_historical_out['Year'].values):
    dato1=0
    for transp_mode in Transport_cross_walk[var_name]:
        dato1+=df_pass.loc[df_pass['Transportation Mode Used on Trip']==transp_mode,'Percent'].values[0]
    
    keys_total=[key for key in Transport_cross_walk.keys() if key.startswith(var_name[0:20])]
    total=0
    for key in keys_total:
        for transp_mode in Transport_cross_walk[key]:
            total+=df_pass.loc[df_pass['Transportation Mode Used on Trip']==transp_mode,'Percent'].values[0]
    
    dato=dato1/total
    df_historical_out.loc[(df_historical_out['iso_code3'] == count) & 
                        (df_historical_out['Year']== year), var_name] = dato
df_historical_out   
```

```python
#### Projected ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years2=range(max(years)+1,2051)

coun_year=[]
for count in zip(countries,ISO3):
    for year in years2:
        coun_year.append([count[0],count[1],year])
df_projected_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
df_projected_out[var_name]=0
for count,year in zip(df_projected_out['Nation'].values,df_projected_out['Year'].values):
    df_dato=df_historical_out[(df_historical_out['Nation']==count) & 
                               (df_historical_out['Year']==max(years))]
    dato=df_dato[var_name].values[0]
    df_projected_out.loc[(df_projected_out['Nation'] == count) & 
                        (df_projected_out['Year']== year), var_name] = dato
df_projected_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
