```python
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)

var_name=os.path.basename(parentdir)

var_name
```

```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
```

```python
df_WEBH2021=pd.read_excel('../row_data/WorldEnergyBalancesHighlights2021.xlsx',sheet_name=3)
```

```python
df_WEBH2021.columns = df_WEBH2021.iloc[0]
df_WEBH2021
```

```python
df_sector=df_WEBH2021[df_WEBH2021['Flow']=='Residential (PJ)']
df_sector
```

```python
nation_in_df=np.unique(df_sector['Country'].values)
nation_in_df
```

```python
df_fuel=df_sector[df_sector['Product']=='Coal, peat and oil shale']
df_fuel.replace(to_replace='..', value=0,inplace=True)
df_fuel
```

```python
df_total=df_sector[df_sector['Product']=='Total']
df_total.replace(to_replace='..', value=0,inplace=True)
df_total
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=[y for y in df_total.columns if type(y)==int]

coun_year=[]
for count in zip(countries,ISO3):
    for year in years:
        coun_year.append([count[0],count[1],year])
df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
df_historical_out[var_name]=0
df_historical_out
```

```python
for count,year in zip(df_historical_out['Nation'].values,df_historical_out['Year'].values):
    if count in nation_in_df:
        Total=df_total.loc[df_total['Country']==count,[year]][year].values[0]
        fuel=df_fuel.loc[df_fuel['Country']==count,[year]][year].values[0]
        
        if Total!=0:
            dato=fuel/Total
            df_historical_out.loc[(df_historical_out['Nation'] == count) & 
                        (df_historical_out['Year']== year), var_name] = dato
        else:
            df_historical_out.loc[(df_historical_out['Nation'] == count) & 
                        (df_historical_out['Year']== year), var_name]=0
            
    else:
        Total=df_total.loc[df_total['Country']=='World',[year]][year].values[0]
        fuel=df_fuel.loc[df_fuel['Country']=='World',[year]][year].values[0]
        dato=fuel/Total
        df_historical_out.loc[(df_historical_out['Nation'] == count) & 
                        (df_historical_out['Year']== year), var_name] = dato
        
    
df_historical_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
#### Projected ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years2=range(max(years)+1,2051)

coun_year=[]
for count in zip(countries,ISO3):
    for year in years2:
        coun_year.append([count[0],count[1],year])
df_projected_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year'])
df_projected_out[var_name]=0
df_projected_out
```

```python
for count,year in zip(df_projected_out['Nation'].values,df_projected_out['Year'].values):
    df_dato=df_historical_out[(df_historical_out['Nation']==count) & 
                               (df_historical_out['Year']==max(years))]
    dato=df_dato[var_name].values[0]
    df_projected_out.loc[(df_projected_out['Nation'] == count) & 
                        (df_projected_out['Year']== year), var_name] = dato
df_projected_out
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
