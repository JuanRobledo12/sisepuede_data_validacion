```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
df_country_continent = pd.read_csv("../row_data/countryContinent.csv",encoding="ISO-8859-1")
df_country_continent.dropna(how='any')
df_country_continent
```

```python
df_WEBH=pd.read_excel("../row_data/WorldEnergyBalancesHighlights2021.xlsx",'TimeSeries_1971-2020')
df_WEBH.columns = df_WEBH.iloc[0]
#df_WEBH.replace(to_replace='..', value=0,inplace=True)
df_WEBH.head()
```

```python
df_reselec = df_WEBH.loc[(df_WEBH['Product']=='Electricity') & (df_WEBH['Flow']=='Residential (PJ)')] 
df_reselec = df_reselec.drop(labels=df_reselec.loc[:, (df_reselec == '..').any()], axis=1)
df_reselec.head()
```

```python
df_popur_hist = pd.read_csv(ggpdir+
        "/SocioEconomic/population_gnrl_urban/input_to_sisepuede/historical/population_gnrl_urban.csv") 
df_popur_proj = pd.read_csv(ggpdir+
        "/SocioEconomic/population_gnrl_urban/input_to_sisepuede/projected/population_gnrl_urban.csv")

df_popru_hist = pd.read_csv(ggpdir+
        "/SocioEconomic/population_gnrl_rural/input_to_sisepuede/historical/population_gnrl_rural.csv") 
df_popru_proj = pd.read_csv(ggpdir+
        "/SocioEconomic/population_gnrl_rural/input_to_sisepuede/projected/population_gnrl_rural.csv")

df_occ_hist = pd.read_csv(ggpdir+
        "/SocioEconomic/occrateinit_gnrl_occupancy/input_to_sisepuede/historical/occrateinit_gnrl_occupancy.csv")
df_occ_proj = pd.read_csv(ggpdir+
        "/SocioEconomic/occrateinit_gnrl_occupancy/input_to_sisepuede/projected/occrateinit_gnrl_occupancy.csv")
```

```python
print(df_popur_hist.head(2))
print(df_popru_hist.head(2))
print(df_occ_hist.head(2))
```

```python
# hh = round((population_rurual + population_urban)/occ_rate)

#countries=df_countries['Country Name'].values
set1 = set(df_popur_hist['iso_code3'].values)
set2 = set(df_popru_hist['iso_code3'].values)
set3 = set(df_occ_hist['iso_code3'].values)

common_set = set1.intersection(set2, set3)
common_elements = sorted(list(common_set))
ISO3=common_elements

#countries=np.unique(df_countries['Nation'].values)
years_hist=range(max(min(zip(df_popur_hist['Year'].values,
                             df_popru_hist['Year'].values,
                             df_occ_hist['Year'].values))),
                min(max(zip(df_popur_hist['Year'].values,
                            df_popru_hist['Year'].values,
                            df_occ_hist['Year'].values)))+1
                )
coun_year=[]
for iso3 in ISO3:
    count = df_countries.loc[df_countries['ISO3']==iso3]['Country Name'].values[0]
    for year in years_hist:
        pop_ru = df_popru_hist.loc[(df_popru_hist['iso_code3'] == iso3) & 
                        (df_popru_hist['Year']== year), 'population_gnrl_rural'].values[0]
        
        pop_ur = df_popur_hist.loc[(df_popur_hist['iso_code3'] == iso3) & 
                        (df_popur_hist['Year']== year), 'population_gnrl_urban'].values[0]
        
        occ_rate = df_occ_hist.loc[(df_occ_hist['iso_code3'] == iso3) & 
                        (df_occ_hist['Year']== year), 'occrateinit_gnrl_occupancy'].values[0]
        
        hh = round((pop_ru+pop_ur)/occ_rate)
        coun_year.append([count,iso3,year,hh])
        
df_hh_hist = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year','hh'])
df_hh_hist
```

```python
years_proj=range(max(min(zip(df_popur_proj['Year'].values,
                             df_popru_proj['Year'].values,
                             df_occ_proj['Year'].values))),
                min(max(zip(df_popur_proj['Year'].values,
                            df_popru_proj['Year'].values,
                            df_occ_proj['Year'].values)))+1
                )
coun_year=[]
for iso3 in ISO3:
    count = df_countries.loc[df_countries['ISO3']==iso3]['Country Name'].values[0]
    for year in years_proj:
        pop_ru = df_popru_proj.loc[(df_popru_proj['iso_code3'] == iso3) & 
                        (df_popru_proj['Year']== year), 'population_gnrl_rural'].values[0]
        
        pop_ur = df_popur_proj.loc[(df_popur_proj['iso_code3'] == iso3) & 
                        (df_popur_proj['Year']== year), 'population_gnrl_urban'].values[0]
        
        occ_rate = df_occ_proj.loc[(df_occ_proj['iso_code3'] == iso3) & 
                        (df_occ_proj['Year']== year), 'occrateinit_gnrl_occupancy'].values[0]
        
        hh = round((pop_ru+pop_ur)/occ_rate)
        coun_year.append([count,iso3,year,hh])
        
df_hh_proj = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year','hh'])
df_hh_proj
```

```python
df_reselec['Country'] = df_reselec['Country'].replace("People's Republic of China", 'China')

# create a list of names to keep
keep_country = df_hh_hist['Nation'].values

# create a boolean mask
mask = df_reselec['Country'].isin(keep_country)

# filter the DataFrame using the mask
df_reselec = df_reselec[mask]

df_reselec.loc[df_reselec['Country']=='China']
```

```python
years1=[c for c in list(df_reselec.columns) if type(c)==int]
years1
mean_WEBH={}
for y in years1:
    mean_WEBH[y]=np.mean(df_reselec[y])
mean_WEBH[2020]=mean_WEBH[2019]
mean_WEBH
```

```python
concinents = ['Australia and New Zealand',
 'Caribbean',
 'Central America',
 'Central Asia',
 'Eastern Africa',
 'Eastern Asia',
 'Eastern Europe',
 'Melanesia',
 'Micronesia',
 'Middle Africa',
 'Northern Africa',
 'Northern America',
 'Northern Europe',
 'Polynesia',
 'South America',
 'South-Eastern Asia',
 'Southern Africa',
 'Southern Asia',
 'Southern Europe',
 'Western Africa',
 'Western Asia',
 'Western Europe']
coun_year=[]
for cont in concinents:
    for year in range(min(years_hist),max(years_hist)):
        coun_year.append([cont,year,0])
        
df_meanWEBH = pd.DataFrame(coun_year, columns=['sub_region', 'Year','meanWEBH'])

for year in range(min(years_hist),max(years_hist)):
    for cont in concinents:
        set1 = set(df_country_continent.loc[df_country_continent['sub_region']== cont]['country'].values)
        set2 = set(df_reselec['Country'].values)

        common_countries = set1.intersection(set2)
        #print(cont,common_countries)
        mean_array=[]
        for count in common_countries:
            mean_array.append(df_reselec.loc[(df_reselec['Country']==count)][year].values[0])
        if mean_array == []:
            mean_array = [0]
        #print(mean_array)
        df_meanWEBH.loc[(df_meanWEBH['sub_region']==cont) & (df_meanWEBH['Year']==year),'meanWEBH']=np.mean(mean_array)
    
df_meanWEBH.dropna(how='any')
df_meanWEBH = df_meanWEBH.drop(df_meanWEBH.index[(df_meanWEBH == 0).any(axis=1)])

df_meanWEBH
```

```python
coun_year=[]
for cont in df_meanWEBH['sub_region'].values:
    for year in range(min(years_proj),max(years_proj)+1):
        m = df_meanWEBH.loc[(df_meanWEBH['sub_region'] == cont) & 
                            (df_meanWEBH['Year']== 2019), 'meanWEBH'].values[0]
        coun_year.append([cont,year,m])
        
df_meanWEBH_proj = pd.DataFrame(coun_year, columns=['sub_region', 'Year','meanWEBH'])
df_meanWEBH_proj
```

```python
#### HISTORICAL ######
countries = set(df_hh_proj['Nation'].values)
coun_year=[]
for count in countries:
    iso3 = df_countries.loc[df_countries['Country Name']==count]['ISO3'].values[0]
    for year in years_hist:
        coun_year.append([count,iso3,year,0])
df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year',var_name])
df_historical_out
```

```python
for c,y in zip(df_historical_out['Nation'].values,df_historical_out['Year'].values):
    if c in df_reselec['Country'].values:
        if y==2020:
            reselec_gj= df_reselec.loc[(df_reselec['Country']==c)][2019].values[0]*1e6
        else:
            reselec_gj= df_reselec.loc[(df_reselec['Country']==c)][y].values[0]*1e6
        hh=df_hh_hist.loc[(df_hh_hist['Nation'] == c) & 
                        (df_hh_hist['Year']== y), 'hh'].values[0]
        dato=reselec_gj/hh
        df_historical_out.loc[(df_historical_out['Nation'] == c) & 
                        (df_historical_out['Year']== y), var_name] = dato
        
    else:
        try:
            cont = df_country_continent.loc[df_country_continent['country']== c]['sub_region'].values[0]
            if y==2020:
                reselec_gj = df_meanWEBH.loc[(df_meanWEBH['sub_region'] == cont) & 
                            (df_meanWEBH['Year']== 2019), 'meanWEBH'].values[0]*1e6
            else:
                reselec_gj = df_meanWEBH.loc[(df_meanWEBH['sub_region'] == cont) & 
                            (df_meanWEBH['Year']== y), 'meanWEBH'].values[0]*1e6
        except:
            reselec_gj=mean_WEBH[y]*1e6
        hh=df_hh_hist.loc[(df_hh_hist['Nation'] == c) & 
                            (df_hh_hist['Year']== y), 'hh'].values[0]
        dato=reselec_gj/hh
        df_historical_out.loc[(df_historical_out['Nation'] == c) & 
                        (df_historical_out['Year']== y), var_name] = dato
df_historical_out
```
