```python
# Add Parent Folder to System Path

import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, parentdir) 

####

import pandas as pd
import numpy as np

#var name
var_name='avgload_trns_freight_tonne_per_vehicle_rail_freight'

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
```

```python
df_countries
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=np.arange(2000,2021,1)
data=np.zeros(len(years))+2923
coun_year=[]
for count in zip(ISO3,countries):
    for year,dat in zip(years,data):
        coun_year.append([year,dat,count[0],count[1]])
df_historical_out = pd.DataFrame(coun_year, columns=['Year',var_name,'iso_code3','Nation'])
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
df_historical_out

```

```python
#### PROJECTED ######

years2=years
data2=data
for y in np.arange(min(years2),2051,1):
    if y not in years2:
        years2=np.append(years2,y)
        data2=np.append(data2,4082)

coun_year2=[]
for count in zip(countries,ISO3):
    for year,dat in zip(years2,data2):
        coun_year2.append([year,dat,count[0],count[1]])

df_projected_out = pd.DataFrame(coun_year2, columns=['Year',var_name,'iso_code3','Nation'])
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
df_projected_out
```
