```python
#Add parent and grandparent folder to sys.path and get variable name as parent folder's name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)


var_name=os.path.basename(parentdir)
var_name
```

```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
```

```python
#### HISTORICAL ######

countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years=np.arange(2000,2023,1)
ppb=np.zeros(len(years))
coun_year=[]
for count in zip(countries,ISO3):
    for year,pb in zip(years,ppb):
        coun_year.append([count[0],count[1],year,pb])
df_historical_out = pd.DataFrame(coun_year, columns=['Country','ISO3', 'Year',var_name])
df_historical_out
```

```python
df_historical_out.to_csv('../input_to_sisepuede/historical/'+var_name+'.csv')
```

```python
#### PROJECTED ######

#Prediction for the next years will be the last historical value
countries=df_countries['Country Name'].values
ISO3=np.unique(df_countries['ISO3'].values)
years2=np.arange(2023,2051,1)
ppb2=np.zeros(len(years2))
coun_year2=[]
for count in zip(countries,ISO3):
    for year,pb in zip(years2,ppb2):
        coun_year2.append([count[0],count[1],year,pb])

df_projected_out = pd.DataFrame(coun_year2, columns=['Country','ISO3', 'Year',var_name])
df_projected_out
```

```python
df_projected_out.to_csv('../input_to_sisepuede/projected/'+var_name+'.csv')
```
