```python
import pandas as pd
import numpy as np

df_countries=pd.read_csv("../row_data/Countries_ISO3.csv")
df_countries
df_country_continent = pd.read_csv("../row_data/iso3_countries.csv")
df_country_continent.dropna(how='any')
df_country_continent
```

```python
df_WEBH=pd.read_excel("../row_data/WorldEnergyBalancesHighlights2021.xlsx",'TimeSeries_1971-2020')
df_WEBH.columns = df_WEBH.iloc[0]
#df_WEBH.replace(to_replace='..', value=0,inplace=True)
df_WEBH.head()
```

```python
df_reselec = df_WEBH.loc[(df_WEBH['Product']=='Electricity') & (df_WEBH['Flow']=='Residential (PJ)')] 
df_reselec = df_reselec.drop(labels=df_reselec.loc[:, (df_reselec == '..').any()], axis=1)
df_reselec.head()
```

```python
df_popur_hist = pd.read_csv("../row_data/population_gnrl_urban_hist.csv") 
df_popur_proj = pd.read_csv("../row_data/population_gnrl_urban_proj.csv")

df_popru_hist = pd.read_csv("../row_data/population_gnrl_rural_hist.csv") 
df_popru_proj = pd.read_csv("../row_data/population_gnrl_rural_proj.csv")

df_occ_hist = pd.read_csv("../row_data/occrateinit_gnrl_occupancy_hist.csv")
df_occ_proj = pd.read_csv("../row_data/occrateinit_gnrl_occupancy_proj.csv")
```

```python
print(df_popur_hist.head(2))
print(df_popru_hist.head(2))
print(df_occ_hist.head(2))
```

```python
# hh = round((population_rurual + population_urban)/occ_rate)

#countries=df_countries['Country Name'].values
set1 = set(df_popur_hist['Nation'].values)
set2 = set(df_popru_hist['Nation'].values)
set3 = set(df_occ_hist['Nation'].values)

common_set = set1.intersection(set2, set3)
common_elements = list(common_set)
countries=common_elements

ISO3=np.unique(df_countries['ISO3'].values)
years_hist=range(max(min(zip(df_popur_hist['Year'].values,
                             df_popru_hist['Year'].values,
                             df_occ_hist['Year'].values))),
                min(max(zip(df_popur_hist['Year'].values,
                            df_popru_hist['Year'].values,
                            df_occ_hist['Year'].values)))+1
                )
coun_year=[]
for count in countries:
    iso3 = df_countries.loc[df_countries['Country Name']==count]['ISO3'].values[0]
    for year in years_hist:
        pop_ru = df_popru_hist.loc[(df_popru_hist['Nation'] == count) & 
                        (df_popru_hist['Year']== year), 'population_gnrl_rural'].values[0]
        pop_ur = df_popur_hist.loc[(df_popur_hist['Nation'] == count) & 
                        (df_popur_hist['Year']== year), 'population_gnrl_urban'].values[0]
        occ_rate = df_occ_hist.loc[(df_occ_hist['Nation'] == count) & 
                        (df_occ_hist['Year']== year), 'occrateinit_gnrl_occupancy'].values[0]
        hh = round((pop_ru+pop_ur)/occ_rate)
        coun_year.append([count,iso3,year,hh])
df_hh_hist = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year','hh'])
df_hh_hist
```

```python
years_proj=range(max(min(zip(df_popur_proj['Year'].values,
                             df_popru_proj['Year'].values,
                             df_occ_proj['Year'].values))),
                min(max(zip(df_popur_proj['Year'].values,
                            df_popru_proj['Year'].values,
                            df_occ_proj['Year'].values)))+1
                )
coun_year=[]
for count in countries:
    iso3 = df_countries.loc[df_countries['Country Name']==count]['ISO3'].values[0]
    for year in years_proj:
        pop_ru = df_popru_proj.loc[(df_popru_proj['Nation'] == count) & 
                        (df_popru_proj['Year']== year), 'population_gnrl_rural'].values[0]
        pop_ur = df_popur_proj.loc[(df_popur_proj['Nation'] == count) & 
                        (df_popur_proj['Year']== year), 'population_gnrl_urban'].values[0]
        occ_rate = df_occ_proj.loc[(df_occ_proj['Nation'] == count) & 
                        (df_occ_proj['Year']== year), 'occrateinit_gnrl_occupancy'].values[0]
        hh = round((pop_ru+pop_ur)/occ_rate)
        coun_year.append([count,iso3,year,hh])
df_hh_proj = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year','hh'])
df_hh_proj
```

```python
years1=[c for c in list(df_reselec.columns) if type(c)==int]
years1
mean_WEBH={}
for y in years1:
    mean_WEBH[y]=np.mean(df_reselec[y])
mean_WEBH[2020]=mean_WEBH[2019]
mean_WEBH
```

```python
concinents = ['AF', 'AN', 'AS', 'EU', 'OC', 'SA']
coun_year=[]
for cont in concinents:
    for year in range(min(years_hist),max(years_hist)):
        coun_year.append([cont,year,0])
        
df_meanWEBH = pd.DataFrame(coun_year, columns=['Continent', 'Year','meanWEBH'])
for year in range(min(years_hist),max(years_hist)):
    for cont in concinents:
        mean=[]
        
        set1 = set(df_country_continent.loc[df_country_continent['Continent']== cont]['Name'].values)
        set2 = set(df_reselec['Country'].values)

        common_countries = set1.intersection(set2)
        
        for count in common_countries:
            mean.append(df_reselec.loc[(df_reselec['Country']==count)][year].values[0])
        #print(cont,year)
        df_meanWEBH.loc[(df_meanWEBH['Year']==year) & (df_meanWEBH['Year']==year),'meanWEBH']=np.mean(mean)
    
df_meanWEBH
```

```python
#### HISTORICAL ######

coun_year=[]
for count in countries:
    iso3 = df_countries.loc[df_countries['Country Name']==count]['ISO3'].values[0]
    for year in years_hist:
        coun_year.append([count,iso3,year,0])
df_historical_out = pd.DataFrame(coun_year, columns=['Nation','iso_code3', 'Year',var_name])
df_historical_out
```
