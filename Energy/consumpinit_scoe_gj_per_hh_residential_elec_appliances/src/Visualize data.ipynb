```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Add parent and grandparent folder to sys.path and get variable name
import os
import sys
import inspect

parentdir = os.path.abspath('..')
grandparentdir=os.path.dirname(parentdir)
ggpdir=os.path.dirname(grandparentdir)

sys.path.insert(0, parentdir)
sys.path.insert(0, grandparentdir)
sys.path.insert(0, ggpdir)
```

```python


var_array=['consumpinit_scoe_gj_per_hh_residential_elec_appliances',
                'consumpinit_scoe_gj_per_hh_residential_heat_energy',
                'consumpinit_scoe_tj_per_mmmgdp_commercial_municipal_elec_appliances',
               'consumpinit_scoe_tj_per_mmmgdp_commercial_municipal_heat_energy',
                'consumpinit_scoe_tj_per_mmmgdp_other_se_elec_appliances',
                'consumpinit_scoe_tj_per_mmmgdp_other_se_heat_energy']

```

```python
# set Countries to be explore

countries=['Brazil','Peru','Mexico','Ecuador','Chile','Argentina','Colombia','Serbia','Montenegro']


df_IEA=pd.read_excel('../row_data/IEA - Energy Efficiency Indicators database - Extended (December 2022).xlsb',
                        engine='pyxlsb',sheet_name='Residential - Energy',skiprows=[0])

# countries in dataframe

countries_in_data=df_IEA['Country'].unique()

# countries status dictionary

countries_dict={}

countries_status=[]

bar_colors=[]



for c in countries:
    if c in countries_in_data:
        status='Observed'
        bar_colors.append('tab:orange')
    else:
        status='Inputed'
        bar_colors.append('tab:green')
    
    countries_dict[c]=status
    
    countries_status.append(c+'-'+status[0:2])
    
countries_dict

countries_status
```

```python
def insert_line_breaks(string, interval):
    return '\n'.join(string[i:i+interval] for i in range(0, len(string), interval))


# fig, axs = plt.subplots(3,round(len(var_array)/3)+1, figsize=(15, 15), facecolor='w', edgecolor='k')
fig, axs = plt.subplots(3,2, figsize=(15, 15), facecolor='w', edgecolor='k')

fig.subplots_adjust(hspace = 1, wspace=0.5)

axs = axs.ravel()

    
for (i,var) in enumerate(var_array):
    countries_new=countries.copy()
    
    df_hist = pd.read_csv(grandparentdir+'/'
                           +var+'/input_to_sisepuede/historical/'
                           +var+'.csv')
    
    # Find the row with the maximum value in the sisepuede_var column
    df_2015=df_hist.loc[(df_hist['Year']==2015)]

    max_row=df_2015.loc[(df_2015[var].idxmax())]
    
#     max_val=max_row.[var].values[0]
    
    max_nation=max_row['Nation']
    
#     countries_new.append(max_nation)

    # Find the row with the minimum value in the sisepuede_var column
    min_row = df_2015.loc[(df_2015[var].idxmin())]
    
    min_nation=min_row['Nation']
    
#     countries_new.append(min_nation)
    
    countries_dict={}

    countries_status=[]

    bar_colors=[]



    for c in countries_new:
        if c in countries_in_data:
            status='Observed'
            bar_colors.append('tab:orange')
        else:
            status='Inputed'
            bar_colors.append('tab:green')

        countries_dict[c]=status

#         countries_status.append(c+'-'+status[0:2])
        if c.startswith(max_nation):
            countries_status.append(c+'-'+status[0:2]+'-Max')
        else:
            if c.startswith(min_nation):
                countries_status.append(c+'-'+status[0:2]+'-Min')
            else:
                countries_status.append(c+'-'+status[0:2])
    values=[]
    for c in countries_new:
        value=df_hist.loc[(df_hist['Nation']==c)&
                         (df_hist['Year']==2015)][var].values[0]
        values.append(value)
        
    axs[i].bar(np.arange(len(countries_new)),values,color=bar_colors)
        
    axs[i].set_title(insert_line_breaks(var,19))
    
    axs[i].set_xticks(np.arange(len(countries_new)),countries_status)
    
#     axs[i].set_xticklabels(countries,rotation=90,color='k',size=10)
    
#     axs[i].xticks(np.arange(len(countries)),countries,rotation=90,color='k',
#            size=5 )

for (i,var) in enumerate(var_array):

    plt.setp(axs[i].get_xticklabels(), rotation=90, ha='right',size=10)
    
```
